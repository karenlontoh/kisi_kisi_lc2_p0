### SQL

##### 1. Berapa banyak label unique dari bike_type?

```python
# Menggunakan DISTINCT untuk mengambil unique value dari kolom 'bike_type'
# Menggunakan COUNT untuk menghitung total unique value dari kolom 'bike_type'
no1 = client.query("""
SELECT
  COUNT(DISTINCT bike_type) AS total_unique_bike_type
FROM ftds-hacktiv8-project.practice.bike_rent;
""").to_dataframe()

no1
```

Tipe sepeda memiliki **2 unique value.**

##### 2. Apa saja nama label unique dari tipe sepeda?

```python
# Menggunakan DISTINCT untuk mengambil nama label unique dari kolom 'bike_type'
no2 = client.query("""
SELECT
  DISTINCT bike_type AS unique_bike_type
FROM ftds-hacktiv8-project.practice.bike_rent;
""").to_dataframe()

no2
```

Tipe sepeda terdiri dari **classic** dan **electric**

##### 3. Berapa banyak total peminjaman sepeda pada tahun 2016 dan 2017?

```python
# Menggunakan COUNT(*) untuk menghitung semua row data peminjaman
# Menggunakan EXTRACT YEAR untuk extract tahun dari kolom 'start_time'
no3 = client.query("""
SELECT
  COUNT(*) AS total_rent
FROM ftds-hacktiv8-project.practice.bike_rent
WHERE EXTRACT(YEAR FROM start_time) IN (2016, 2017);
""").to_dataframe()

no3
```

Total peminjaman sepeda pada tahun 2016 sampai 2017 adalah **7,993 peminjaman**

##### 4. Dimana station yang paling banyak dikunjungi sebagai titik keberangkatan pada tahun 2016?

```python
# Menggunakan EXTRACT YEAR untuk extract tahun dari kolom 'start_time'
# Menggunakan COUNT untuk menghitung banyak peminjaman
# Menggunakan ORDER BY untuk sorting mulai dari yang terbesar (descending)
# Menggunakan LIMIT untuk mengambil row pertama yang merupakan data terbesar
no4 = client.query("""
SELECT
  start_station_name
FROM ftds-hacktiv8-project.practice.bike_rent
WHERE EXTRACT(YEAR FROM start_time) = 2016
GROUP BY start_station_name
ORDER BY COUNT(start_station_name) DESC
LIMIT 1;
""").to_dataframe()

no4
```

Stasiun **Bullock Museum @ Congress & MLK** merupakan stasiun yang paling sering dijadikan titik keberangkatan oleh peminjam pada tahun 2016

##### 5. Berapa nilai minimum, maximum, dan rata-rata dari durasi perjalanan pada keseluruhan data?

```python
# Menggunakan MIN untuk mendapatkan nilai minimum durasi perjalanan
# Menggunakan MAX untuk mendapatkan nilai maximum durasi perjalanan
# Menggunakan AVG untuk mendapatkan nilai average durasi perjalanan
no5 = client.query("""
SELECT
  MIN(duration_minutes) AS minimum_durasi,
  MAX(duration_minutes) AS maximum_durasi,
  AVG(duration_minutes) AS rata_rata_durasi
FROM ftds-hacktiv8-project.practice.bike_rent;
""").to_dataframe()

no5
```

Peminjaman sepeda pada tahun 2016 sampai 2018 memiliki waktu peminjaman paling singkat **2 menit** dan paling lama selama **2,218 menit**. Rata-rata peminjam melakukan peminjaman sepeda selama **35.09 menit**

##### 6. Berapa rata-rata durasi perjalanan dari subscriber_type `Explorer` dan `Weekender`?

```python
# Menggunakan AVG untuk mendapatkan nilai average durasi perjalanan
no6 = client.query("""
SELECT
  subscriber_type,
  AVG(duration_minutes) as rata_rata_durasi
FROM ftds-hacktiv8-project.practice.bike_rent
WHERE subscriber_type IN ('Explorer','Weekender')
GROUP BY subscriber_type;
""").to_dataframe()

no6
```

Peminjam dengan tipe `Explorer` untuk peminjaman tahun 2016 sampai 2018 rata-rata melakukan peminjaman selama **42.86 menit** sedangkan peminjam dengan tipe `Weekender` rata-rata melakukan peminjaman selaman **27.63 menit**.

##### 7. Tampilkan entry data yang memiliki `duration_time` diatas 4.25 jam.

```python
# Subquery untuk mengubah data 'duration_minutes' dari satuan menit ke satuan jam
no7 = client.query("""
SELECT *
FROM (
    SELECT *,
    (duration_minutes / 60) AS duration_time
    FROM ftds-hacktiv8-project.practice.bike_rent
    )
WHERE duration_time > 4.25;
""").to_dataframe()

no7
```

Data di atas merupakan data peminjaman dengan durasi pinjam lebih dari **4.25 jam**

### Pandas

```python
# Load data frame
df = client.query("""
SELECT *
FROM ftds-hacktiv8-project.practice.bike_rent;
""").to_dataframe()

# Menampilkan data frame
df
```

##### 1. Cek kesesuaian tipe data dan keberadaan missing values dengan `.info()`

```python
# Menampilkan informasi tabel dengan `.info()`
df.info()
```

Berdasarkan informasi tabel, dapat dilihat bahwa:
1. Terdapat **tipe data yang salah**, yaitu tipe data kolom `bike_id` yang seharusnya integer.
2. **Tidak terdapat missing values**, ditandai dengan informasi bahwa seluruh kolom memiliki 8,000 non-null dari total 8,000 entries

##### 2. Buat kolom baru bernama `quarter`.

```python
# Inisiasi insance untuk define month dari kolom `start_time`
month_num = df['start_time'].dt.month

# Inisiasi list untuk menyimpan data kuartal
kuartal = []

# Loop untuk menentukan kuartal berdasarkan bulan
for month in month_num:
    if month in [1, 2, 3]:
        kuartal.append('Q1')
    elif month in [4, 5, 6]:
        kuartal.append('Q2')
    elif month in [7, 8, 9]:
        kuartal.append('Q3')
    elif month in [10, 11, 12]:
        kuartal.append('Q4')

# Membuat kolom baru `quarter` dengan hasil kuartal
df['quarter'] = kuartal

# Menampilkan data frame yang sudah ditambahkan kolom `quarter`
df
```

Sudah terdapat kolom baru, yaitu kolom `quarter` yang menjelaskan tentang peminjaman dilakukan di kuarter ke berapa.

##### 3. Tampilkan data perjalanan dari tipe member `Local31` dengan durasi perjalanan diatas 1 jam.


```python
# Menampilkan data perjalanan dengan filtering `subscriber_type` = Local31 dan `duration_minutes` > 60 menit
df[(df['subscriber_type'].str.contains('Local31')) & (df['duration_minutes'] > 60)]
```

**Tidak terdapat data** peminjaman oleh member `Local31` yang melakukan peminjaman selama lebih dari 1 jam

##### 4. Berapa rata-rata durasi perjalanan dari tipe pelanggan `Walk Up` dengan tipe sepeda `classic` pada bulan `Maret` dari data yang diambil?


```python
# Filtering data untuk kondisi `subscriber_type` = Walk Up, `bike_type` = classic, dan `start_time` = 3 (bulan Maret)
no4 = df[ (df['subscriber_type'] == 'Walk Up') &
          (df['bike_type'] == 'classic') &
          (df['start_time'].dt.month == 3)]

# Menampilkan rata-rata (mean) dari `duration_minutes` berdasarkan data yang sudah dilakukan filtering
no4['duration_minutes'].mean()
```

Rata-rata peminjam `Walk Up` yang meminjam sepeda `classic` pada bulan `Maret`, melakukan peminjaman selama **38.47 menit**
